### Tag document

```python
i = 0
train_data_doc = []
for list_t in train_data['title_common_noun_words']:
    res = TaggedDocument(list_t, i)
    train_data_doc.append(res)
    i = i + 1
```

```python
train_data_doc[2:10]
```

## Train model

```python
max_epochs = 100
vec_size = 20
alpha = 0.025
```

```python
model = Doc2Vec(alpha=alpha, 
                min_alpha=0.00025,
                min_count=1,
                dm =1)
```

```python
model.build_vocab(train_data_doc)
```

```python
model.train(train_data_doc, total_examples=model.corpus_count, epochs=model.epochs)
```

### Ranks

```python
ranks = []
second_ranks = []
for doc_id in range(len(train_data_doc)):
    inferred_vector = model.infer_vector(train_data_doc[doc_id].words)
    sims = model.docvecs.most_similar([inferred_vector], topn=len(model.docvecs))
    rank = [docid for docid, sim in sims].index(doc_id)
    ranks.append(rank)
    
    second_ranks.append(sims[1])
```
